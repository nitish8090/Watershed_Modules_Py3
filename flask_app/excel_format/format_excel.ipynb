```python
df = pd.read_excel("2nd_part.xlsx")
```

```python
df
```

```python
list_items = []
for column_head, column_data in df.iteritems():
    for i, column_entry in enumerate(column_data):
        if i >1:
            if not pd.isnull(column_entry):
                village_name = column_entry.split("(")[0]
                village_code = column_entry.split("(")[-1][:-1]
                
                result_statement = "{},{},{},{}".format(column_data[0], column_data[1], village_name, village_code)
                list_items.append(result_statement)
    print("_________")
```

```python
with open("result.csv", "a") as result_csv:
    for item in list_items:
        result_csv.write(item)
        result_csv.write("\n")
```
